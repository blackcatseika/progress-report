```python
import json
import pandas as pd
```

```python
def best_within_one_1(target_model_id, target_model_type):
    # load trials data
    trials_file_path = f'results/{target_model_id}_{target_model_type}_trials.json'
    df_trials = pd.json_normalize(
        [json.loads(line) for line in open(trials_file_path, 'r', encoding='utf-8')]
    )
    # load best data
    best_trial_path = f'results/{target_model_id}_{target_model_type}_trial_best.json'
    with open(best_trial_path, 'r', encoding='utf-8') as f:
        top_performance = json.load(f)
    # calculate 1-se performance floor
    performance_floor = top_performance['value'] - top_performance['user_attrs']['std_err']
    # select all candidate models
    candidate_models = df_trials[df_trials['value'] >= performance_floor].copy()
    # define the priority order for sorting model complexity (based on parameter importance)
    complexity_ranking = [
        "params.learning_rate",
        "params.subsample",
        "params.max_depth"
    ]
    # take first line 
    most_parsimonious_model = candidate_models.sort_values(by=complexity_ranking).iloc[0]
    # extract and save final result
    final_params = most_parsimonious_model.filter(like='params.')
    final_params.index = [col.replace('params.', '') for col in final_params.index]
    final_params['achieved_value'] = most_parsimonious_model['value']

    output_file = f"results/{target_model_id}_{target_model_type}_best_within_one.json"
    final_params.to_json(output_file, orient='index', indent=4)
```

```python
best_within_one_1("model_1", "xgboost")
```
