```python
import json
import pickle
import numpy as np
import pandas as pd
```

```python
count = "02"

model_number = "model_2"

model_types = ["logistic_reg",
               "binary_tree",
               "xgboost",
               "rf"]
```

```python
results = []
for type in model_types:
    file_path = f"results/{model_number}_{type}.pkl"
    with open(file_path, "rb") as f:
        cv_data = pickle.load(f)
    n_folds = cv_data["test_precision"].size
    metrics_summary = {
        "model_number": model_number,
        "model_type": type,
        "precision": cv_data["test_precision"].mean(),
        "precision_std": cv_data["test_precision"].std() / np.sqrt(n_folds),
        "recall": cv_data["test_recall"].mean(),
        "recall_std": cv_data["test_recall"].std() / np.sqrt(n_folds),
        "roc_auc": cv_data["test_roc_auc"].mean(),
        "roc_auc_std": cv_data["test_roc_auc"].std() / np.sqrt(n_folds),
        "accuracy": cv_data["test_accuracy"].mean(),
        "accuracy_std": cv_data["test_accuracy"].std() / np.sqrt(n_folds),
        "f1": cv_data["test_f1"].mean(),
        "f1_std": cv_data["test_f1"].std() / np.sqrt(n_folds)
    }
    results.append(metrics_summary)
```

```python
final_report = pd.DataFrame(results)
output_filename = f"results/{count}_{model_number}_final_results.csv"
final_report.to_csv(output_filename, index=False)
```
